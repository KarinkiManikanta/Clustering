```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_profiling
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/reddyprasade/Machine-Learning-Problems-DataSets/master/Clustering/Wholesale%20customers%20data.csv")
```

```python
pandas_profiling.ProfileReport(data)
```

```python
from sklearn.preprocessing import normalize
data_scaled=normalize(data)
data_scaled=pd.DataFrame(data_scaled,columns=data.columns)
data_scaled.head()
```

```python
sns.distplot(data_scaled)
```

```python
pandas_profiling.ProfileReport(data_scaled)
```

# Normalized Data

```python
# ward linkage
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(30,10))
plt.title("Dendrogram_ward")
ward=shc.dendrogram(shc.linkage(data_scaled,method='ward'))
```

```python
# average linkage
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(30,10))
plt.title("Dendrogram_average")
ward=shc.dendrogram(shc.linkage(data_scaled,method='average'))
```

```python
# complete linkage
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(30,10))
plt.title("Dendrogram_complete")
ward=shc.dendrogram(shc.linkage(data_scaled,method='complete'))
```

```python
# single linkage
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(30,10))
plt.title("Dendrogram_single")
ward=shc.dendrogram(shc.linkage(data_scaled,method='single'))
```
